Schleife zur Bildanpassung 

```python
for i in range(0, len(X)):
    frame = X[i]
    image = cv2.imread(frame)
    
    cropped = image[200:800, :, :]
  
    resize = cv2.resize(cropped, (200, 66), cv2.INTER_AREA) 
    
    yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
 
    cv2.imshow("Bild", image)
    cv2.imshow("Resized", resize)    
    cv2.imshow("Yuv", yuv)
    
    cv2.waitKey(0)
```
